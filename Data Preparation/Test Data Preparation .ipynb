## Test Data Preparation

```python
import pandas as pd
import re
import csv
```

```python
# Function to preprocess data
def preprocess(str):
    re.sub(r'[^\x00-\x7F]+',' ', str)
    tmp="".join([i if ord(i) < 128 else ' ' for i in str])
    ans=" ".join(tmp.split())
    ans=ans.replace('- ',' ')
    ans=ans.replace('? ',' ')
    return ans
```

```python
# Reading test data
df=pd.read_csv('testx.csv',delimiter=",")
```

```python
df
```

```python
# preprocess description
for i in range(len(df)):
    df.iat[i,0]=preprocess(df.iat[i,0])
```
