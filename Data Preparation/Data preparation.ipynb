```python
classes=[['Software Development Manager','IT Manager','Project Manager'],['Architect'],['Lead'],['Quality Assurance','Tester','QA'], \
         ['Software Engineer','Developer','Programmer'],['Database']]
```

```python
def changeClasses(listx):
    newList=[]
    print(len(listx))
    for i in range(0,len(listx)):
        flag=False
        j=0
        while j<6:
            for str in classes[j]:
                if str in listx[i][1]:
                    flag=True
            if flag==True:
                newList.append([listx[i][0],finalClasses[j]])
                break
            j=j+1
    return newList
```

```python
listJR=changeClasses(listJR)
```

```python
def findClassFrequency(listx,classname):
    l=0
    for i in listx:
        if i[1]==classname:
            l=l+1
    return l
```

```python
def finalDataList(listx,maxcount):
    arr=[]
    for classname in finalClasses:
        count=0
        for desc,title in listx:
            if title==classname:
                arr.append([desc,title])
                count=count+1
            if count==maxcount:
                break
    return arr
```

```python
def preprocess(str):
    re.sub(r'[^\x00-\x7F]+',' ', str)
    tmp="".join([i if ord(i) < 128 else ' ' for i in str])
    ans=" ".join(tmp.split())
    ans=ans.replace('- ',' ')
    return ans
```

```python
preprocess("- Rendering technical assistance to Database Management Systems;\r\n- Realization of SQL servers maintenance activities: back-up and\r\nreplication;\r\n- Participation in designing of software development projects.")
```
